```python
#Histogram plotting

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter

# Fixing random state for reproducibility
np.random.seed(19680801)

N_points = 100000
n_bins = 20 
# N_bin adalah pembagian, jika n_bin nya ada 5 dan data nya ada 100 maka n_bin nya adalah kelipatan 20 40 60 80 100

# Generate a normal distribution, center at x=0 and y=5 (Symetric function)
x = np.random.randn(N_points)
y = .4 * x + np.random.randn(100000) + 5 
#0.4 * nilai random + nilai random dengan range 100000 + 5

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
# Subplot (1 baris 2 kolom berarti grafik nya kesamping)


# We can set the number of bins with the `bins` kwarg
axs[0].hist(x, bins=n_bins)
axs[1].hist(y, bins=n_bins)
```

```python
fig, axs = plt.subplots(1, 2, tight_layout=True)

# N is the count in each bin, bins is the lower-limit of the bin
N, bins, patches = axs[0].hist(x, bins=n_bins)

# We'll color code by height, but you could use any scalar
fracs = N / N.max()

# we need to normalize the data to 0..1 for the full range of the colormap
norm = colors.Normalize(fracs.min(), fracs.max())

# Now, we'll loop through our objects and set the color of each accordingly
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)

# We can also normalize our inputs by the total number of counts
axs[1].hist(x, bins=n_bins, density=True)

# Now we format the y-axis to display percentage
axs[1].yaxis.set_major_formatter(PercentFormatter(xmax=1))
```

```python
#Bar charts

plt.rcdefaults()

objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))

performance = [10,8,6,4,2,1]

plt.bar(y_pos, performance, align = 'center', alpha = 0.5)
plt.xticks (y_pos, objects) # Untuk Label pada X
plt.ylabel('Usage')
plt.title('Programming Language Usage')

plt.show()
```

```python
#Waffle diagram

import matplotlib.pyplot as plt
from pywaffle import Waffle
# The values are rounded to 10 * 5 blocks
fig = plt.figure(
    FigureClass=Waffle, 
    rows=5, 
    columns=10, 
    values=[48, 46, 3]
)
plt.show()
```

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle
data = {'Democratic': 48, 'Republican': 46, 'Libertarian': 3}
fig = plt.figure(
    FigureClass=Waffle, 
    rows=5, 
    values=data, 
    legend={'loc': 'upper left', 'bbox_to_anchor': (1.1, 1)}
)
plt.show()
```
